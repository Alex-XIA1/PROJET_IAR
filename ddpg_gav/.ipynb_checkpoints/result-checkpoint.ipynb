```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt("./resultats/resultat.txt")
y = data.mean(0)
y = np.hstack((0,y))
x = np.arange(0,101)*1024
```

```python
for y in data[:5]:
    y = np.hstack((0,y))
    x = np.arange(0,101)*1024
    plt.plot(x,y)
    
plt.xlabel("steps")
plt.ylabel("reward")
plt.title("Performances Gav (max 500), gamma = 0.95, tanH (action entre -1 et 1)")
plt.show()
```
